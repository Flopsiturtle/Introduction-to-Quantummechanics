```python
import numpy as np
%display latex

Psi = var('Psi')
x = var('x')
A, a, b = var('A', 'a', 'b')
a_val = 2
b_val = 3
A_val = sqrt(3/b_val)

assume(A>0, a>0, b>0)

eq0 = Psi == piecewise([((-oo,0), 0), ((0,a_val),A_val*(x/a_val)), ([a_val, b_val],A_val*(b_val-x)/(b_val-a_val)), ((b_val,+oo),0)], var=x)
eq1 = eq0.rhs()*eq0.rhs().conjugate() == eq0.lhs()*eq0.lhs()
show(eq0)
P = plot(eq0.lhs(),(-5,5), color = 'blue', legend_label=r"$%s$"%latex(eq0.rhs()), axes_labels=['x','y']) + plot(eq0.lhs()*eq0.lhs().conjugate(),(-5,5), color = 'orange', legend_label=r"$%s$"%latex(eq1.lhs()))
P
```

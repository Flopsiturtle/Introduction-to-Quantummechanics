# Infinite Square Well

```python
#height of the line for the sketch of the potential
h = 2
#x coordinate of right side of the Infinite square well
a = 1

#two vertical lines one from point (a,0) to (a,h) and the other from point (0,0) to (0,h)
V = line([(a,0), (a,h)], color = 'black', legend_label = 'V') + line([(0,0), (0,h)], color = 'black')
```

```python
#Function of Problem 2.7
A = sqrt(12/a^3)
f = piecewise([([0,a/2], A*x), ((a/2, a), A*(a-x))], var = x)

#plotting function and function squared
F = plot(f, (0,a), color = 'orange', legend_label = 'Psi') + plot(f*f, (0,a), color = 'red', legend_label = '|Psi|^2')
V + F
```

```python
#Function of Problem 2.8
A = sqrt(2/a)
f = piecewise([([0,a/2], A), ((a/2, a), 0)], var = x)
#plotting function and function squared
F = plot(f, (0,a), color = 'orange', legend_label = 'Psi') + plot(f*f, (0,a), color = 'red', legend_label = '|Psi|^2')
V + F
```
